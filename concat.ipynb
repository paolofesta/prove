```python
import numpy as np
import pandas as pd
```

```python
cgil=pd.read_csv('/content/cgil_data')
cgil['sigla']=np.array('CGIL')
```

```python
cisl=pd.read_csv('/content/cisl_data')
cisl['sigla']=np.array('CISL')
```

```python
uil=pd.read_csv('/content/uil_data')
uil['sigla']=np.array('UIL')
```

```python
data=pd.concat([cgil,cisl,uil],ignore_index= True)
data
```

```python
data.to_csv('data')
```
